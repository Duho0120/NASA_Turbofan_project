```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 인코딩 문제 해결
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지
```

```python
RUL_01 = pd.read_csv('../behrad3d/nasa-cmaps/versions/1/CMaps/RUL_FD001.txt', header=None, sep=' ')
train_01 = pd.read_csv('../behrad3d/nasa-cmaps/versions/1/CMaps/train_FD001.txt', header=None, sep=' ').dropna(axis=1)
test_01 = pd.read_csv('../behrad3d/nasa-cmaps/versions/1/CMaps/test_FD001.txt', header=None, sep=' ').dropna(axis=1)

col_names = ['unit', 'cycle', 'setting_1', 'setting_2', 'setting_3', 'T2', 'T24', 'T30', 'T50', 'P2', 'P15', 'P30', 'Nf', 'Nc', 'epr', 'Ps30', 'phi', 'NRf', 'NRc', 'BPR', 'farB', 'htBleed', 'Nf_dmd', 'PCNfR_dmd', 'W31', 'W32']

train_01.columns = col_names
test_01.columns = col_names
```

```python
train_01.head()
```

```python
unit1 = train_01[train_01['unit'] == 1]

# 각 센서 데이터의 변화
plt.figure(figsize=(15, 20))
for i, sensor in enumerate(col_names[5:]):  # 센서 데이터는 5번째 열부터 시작
    plt.subplot(8, 3, i + 1)
    sns.lineplot(data=unit1, x='cycle', y=sensor)
    plt.title(f'Unit 1 - {sensor} over Cycles')
    plt.xlabel('Cycle')
    plt.ylabel(sensor)
plt.tight_layout()
plt.show()
```

```python
unit2 = train_01[train_01['unit'] == 2]

# 각 센서 데이터의 변화
plt.figure(figsize=(15, 20))
for i, sensor in enumerate(col_names[5:]):  # 센서 데이터는 5번째 열부터 시작
    plt.subplot(8, 3, i + 1)
    sns.lineplot(data=unit2, x='cycle', y=sensor)
    plt.title(f'Unit 2 - {sensor} over Cycles')
    plt.xlabel('Cycle')
    plt.ylabel(sensor)
plt.tight_layout()
plt.show()
```

```python
# 사이클에 따른 환경 조건 변화
unit1 = train_01[train_01['unit'] == 6]
plt.figure(figsize=(12, 8))
for i, setting in enumerate(col_names[2:5]):  # 환경 설정은 2~4번째 열
    plt.subplot(3, 1, i + 1)
    sns.lineplot(data=unit1, x='cycle', y=setting)
    plt.title(f'Unit 1 - {setting} over Cycles')
    plt.xlabel('Cycle')
    plt.ylabel(setting)
plt.tight_layout()
plt.show()
```

사이클이 증가함에 따라 각 센서값의 변화를 시각화 하였으며 잡음이 존재하는 것을 알 수 있음

```python
# 모든 피쳐들의 표준편차 확인
feature_std = train_01.iloc[:, 2:].std()
plt.figure(figsize=(12, 6))
sns.barplot(x=feature_std.index, y=feature_std.values)
plt.xticks(rotation=90)
plt.title('Feature Standard Deviations')
plt.xlabel('Features')
plt.ylabel('Standard Deviation')
plt.show()
```

센서 값의 변화는 T30, T50, Nc, NRc 등이 뚜렷하게 나타남

```python
# 사이클에 따른 환경 조건 변화
unit1_04 = train_04[train_04['unit'] == 2]
plt.figure(figsize=(12, 8))
for i, setting in enumerate(col_names[2:5]):  # 환경 설정은 2~4번째 열
    plt.subplot(3, 1, i + 1)
    sns.lineplot(data=unit1_04, x='cycle', y=setting)
    plt.title(f'Unit 1 - {setting} over Cycles')
    plt.xlabel('Cycle')
    plt.ylabel(setting)
plt.tight_layout()
plt.show()
```

```python
# 각 센서들에 대한 LOWESS 시각화
from statsmodels.nonparametric.smoothers_lowess import lowess

unit1_04 = train_04[train_04['unit'] == 1]
plt.figure(figsize=(12, 8))
for i, sensor in enumerate(col_names[5:]):  # 센서 데이터는 5번째 열부터 시작
    plt.subplot(8, 3, i + 1)
    lowess_smoothed = lowess(unit1_04[sensor], unit1_04['cycle'], frac=0.1) # frac는 부드러움 정도 조절
    plt.plot(lowess_smoothed[:, 0], lowess_smoothed[:, 1], color='blue') # LOWESS 곡선
    plt.xlabel('Cycle')
    plt.ylabel(sensor)
# 수직 간격 조정
plt.tight_layout(h_pad=1)
# y 간격이 너무 작으니 조정
plt.subplots_adjust(hspace=0.5)
plt.show()
```
